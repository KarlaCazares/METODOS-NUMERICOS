```python
import numpy as np
def secante(f,x0,x1,nmax,tol):
    for i in range(0,nmax):
        x2=(x0*f(x1)-x1*f(x0))/(f(x1)-f(x0));
        if abs(x2-x1) < tol:
            print(x2,' encontrado con ',i,' interacciones')
            break
        else:
            x0=x1
            x1=x2
    return None

#f = lambda x: x*np.cos(x)-3
f=lambda x: x+np.cos(x)
secante(f,-2,0,20,1e-10)
```

#### Consideramos un sistema de3 ecuaciones y 3 incognitas como los describimos:
$$
\begin{array}{r}
a_{11} x_{1}+a_{12 x_{2}}+a_{13} x_{3}=b_{1} \\
a_{22} x_{2}+a_{23} x_{3}=b_{2} \\
a_{33} x_{3}=b_{3}

$$
Para hallar la solucin podemos despejar $x_{3}$ en la ltima ecuacin:
$$
x_{3}=\frac{b_{3}}{a_{33}}
$$
Con el valor $x_{3}$ despejar $x_{2}$
$$
x_{2}=\frac{b_{2}-a_{23} x_{3}}{a_{22}}
$$
Con los valores hallados $x_{3} x_{2}$, despejar $x_{1}$
$$
x_{1}=\frac{b_{1}-a_{12} x_{2}-a_{13} x_{3}}{x_{11}}
$$
#### Algoritmo de sustitucion regresiva

$$x_{i}=\frac{b_{i}-\sum_{j=i+1}^{n} a_{i j} x_{j}}{\text { aii }}
$$
#### Ejemplo:
$$
\begin{aligned}
x+2 y-2 z &=5 \\
y-z &=2 \\
z &=0
\end{aligned}
$$

#### Vamos a emplear este metodo para poder resolver sistemas lineales de n ecuaciones y n incognitas donde la matrizde coeficientes tiene ceros debajo de la diagonal.

- Forma matricial.

$$
\left(\begin{array}{ccc}
a_{11} & a_{12} & a_{13} \\
0 & a_{22} & a_{23} \\
0 & 0 & a_{33}
\end{array}\right)\left(\begin{array}{l}
x_{1} \\
x_{2} \\
x_{3}
\end{array}\right)=\left(\begin{array}{l}
b_{1} \\
b_{2} \\
b_{3}
\end{array}\right)
$$
- Forma aumentada:
$$
\left(\begin{array}{cccc}
a_{11} & a_{12} & a_{13} & \mid b_{1} \\
0 & a_{22} & a_{23} & \mid b_{2} \\
0 & 0 & a_{33} & \mid b_{3}
\end{array}\right)
$$
### Codigo

```python
### Codigo de la matriz
import numpy as np
# Matriz del sistema A
A=np.array([[1,2,-2],[0,1,-1],[0,0,1]])
#Terminos independientes
b=np.array([5,2,0])
#Hallamos el numero de ecuaciones
n=np.size(b)
#Define un vector para almacenar la solucion
x=np.zeros(n)
#Este bucle va air desde la ultima ecuacion hasta la primera
for i in range (n-1,-1,-1):
    sumj=0
    for j in range (i+1,n):
        sumj+=A[i,j]*x[j]
    x[i]=(b[i]-sumj)*1/A[i,i]

print(x)
            
```

### 5. Programe el pivoteo parcial en el algoritmo de eliminacin gaussiana. Utilice este nuevo programa para resolver el ejercicio anterior. Compare los resultados.
#### Codigo

```python
import numpy as np
# Ingresamos los datos anteriores
A = np.array([[4, -1, 0, -1],
              [-1, 4, -1, 0],
              [0, -1, 4, -1],
              [-1, 0, -1, 4]])
B = np.array([[30],
              [60],
              [70],
              [40]])
# Procedemos a resolver
# El pivoteo nos permite ver la filas consideradas en cero
casi_cero = 1e-15 
# Nos sirve para evitar el truncamiento en operaciones
A = np.array(A, dtype=float)
AB = np.concatenate((A, B), axis=1)
# Pivoteo por filas
tam = np.shape(AB)
n = tam[0]
m = tam[1]

for i in range(0, n - 1, 1):
    columna = abs(AB[i:, i])
    donde_max = np.argmax(columna)
    if (donde_max != 0):
# Intercambiamos las filas
        temporal = np.copy(AB[i, :])
        AB[i, :] = AB[donde_max + i, :]
        AB[donde_max + i, :] = temporal
AB1 = np.copy(AB)
#Eliminacion Gausiana con pivoteo
for i in range(0, n - 1, 1):
    pivote = AB[i, i]
    adelante = i + 1
    for k in range(adelante, n, 1):
        factor = AB[k, i] / pivote
        AB[k, :] = AB[k, :] - AB[i, :] * factor
# sustituimos
ult_fila = n - 1
ult_columna = m - 1
# variable x
X = np.zeros(n, dtype=float)
for i in range(ult_fila, 0 - 1, -1):
    suma = 0
    for j in range(i + 1, ult_columna, 1):
        suma = suma + AB[i, j] * X[j]
    B1 = AB[i, ult_columna]
    X[i] = (B1 - suma) / AB[i, i]
X = np.transpose([X])
# impresion
print("\x1b[1;28m"+"Metodo de Pivoteo Parcial Por Filas")
print(AB1)
print("\x1b[1;28m"+"La Solucin del ejercicio es: ")
print(X)
```

##### Puedo concluir que el las tcnicas de pivoteo, dicta que el elemento pivote que debe escogerse al mayor de cada columna. ,as inicialmente, se organiza la matriz para que el elemento absolutamente mayor de la primera columna sea el primer pivote, luego se realiza la eliminacin y a la hora de pasar al segundo elemento debe considerarse el mayor absolutamente para ubicarlo como pivote y nuevamente hacer la eliminacin, el proceso sigue hasta que ya no hayan elementos bajo el pivote o hasta que se hayan recorrido todas las columnas de la matriz. Este mtodo aun teniendo errores de redondeo o de propagacin son menores que los que se presentan en el mtodo de eliminacin gaussiana.
