### 2.- Desarrolle en Python un programa para calcular la inversa de matrices de dimensin  2x2  No olvide colocar comentarios en su programa. No busque el programa en internet.

```python
### Desarrollo en Python para calcular la inversa de matrices de dimension 2x2.

print("Ingreso de valores para una matriz de 2x2") #ingresamos los valores para la m
a=float(input("Ingrese el valor de a:"))
b=float(input("Ingrese el valor de b:"))
c=float(input("Ingrese el valor de c:"))
d=float(input("Ingrese el valor de d:"))
#calculamos el determinante de la matriz
detA=a*d-b*c #operacion para calcular la determinante de una matriz de 2x2
if (detA==0): #condicional para comprobar que la matriz tiene inversa ya que no existe
 print("La matriz no tiene inversa") #imprimimos el mensaje que notifica que la matriz no tiene inversa
else:
 print("EL determinante de A es :",detA) #si existe realizamos el calculo de la determinante
#calculamos la inversa de la matriz
print("La matriz inversa es:")
a1=(1/detA)*(d) #realizamos los calculos de los elementos de la matriz
b1=(1/detA)*(-b)
c1=(1/detA)*(-c)
d1=(1/detA)*(a)


#Imprimimos la matriz inversa
print("[",a1," ", b1,"]")
print("[",c1," ", d1,"]")
```

### 6.- Disee un cdigo que encuentre el $\sin{\frac{\pi}{3}}$ a travs del desarrollo de Taylor, truncar cuando $n=50$
- Serie de Taylor:
$$(x)=\sum\limits_{k=0}^\infty {\frac{f^{k}(x_0)}{k!}(x-x_0)^{k}}$$
- Serie de Taylor expandida:
$$f(x)=f(x_0)+\frac{f'(x_0)}{1!}(x-x_0)+\frac{f''(x_0)}{2!}(x-x_0)^{2}+\frac{f'''(x_0)}{3!}(x-x_0)^{3}+.....+$$


- Desarrollo analitico de $\sin(x)$ y $x_0=0$:


$\begin{array}{cc}
                f^{k}(x_0) &&&&&&&& f^{k}(0) \\
                f(x)=\sin(x) &&&&&&&& f(0)=\sin{0}=0 \\
                f'(x)=\cos(x) &&&&&&&& f'(0)=\cos(0)=1 \\
                f''(x)=-\sin(x) &&&&&&&& f''(0)=-\sin(0)=0
  \end{array}$
  
 - Serie de taylor del Seno de $\bar{X}$
 
$Sen(X)=0+\frac{1}{1 !} x+\frac{0}{21} x^{2}+\frac{-1}{3 !} x^{3}+\frac{0}{4} x^{4}+\frac{1}{5 !} x^{5}+\frac{0}{6 !} x^{6}+\frac{-1}{71} x^{7}+\frac{0}{8 !} x^{8}+\frac{1}{9 !} x^{9}$

$Sen(x)=-\frac{1}{1 !} x-\frac{1}{3 !} x^{3}+\frac{1}{61} x^{5}-\frac{1}{7 !} x^{7}+\frac{1}{9 !} x^{9}+\ldots$
#### Modelamiento de la Serie
$\sum_{n=1}^{\infty}(-1)^{n+1} \frac{x^{2 n-1}}{(2 n-1) !}$

```python
## codigo de la serie
import numpy as pn
x=pn.pi/3
n=51
polinomio=0
print()
print("n  Pn(x)")
for k in range(n):
    polinomio=polinomio + (-1)**k*x**(2*k+1) /pn.math.factorial(2*k+1)
    print(k,polinomio)
```
