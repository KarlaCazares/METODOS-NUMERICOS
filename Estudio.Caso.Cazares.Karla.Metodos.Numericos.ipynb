```python
import numpy as np
def tridiag(n,d,a,c,b):
    x = np.zeros(n)
    for i in range(1,n-1):
        mult = a[i-1]/d[i-1]
        d[i] -= mult*c[i-1]
        b[i] -= mult*b[i-1]
    print(d)
    print(b)
    # Sustitucion regresiva
    x[n-1] = b[n-1]/d[n-1]
    for i in range(n-2,-1,-1):
        x[i] = (b[i]-c[i]*x[i+1])*d[i]
        #print(x)
    return x
```
