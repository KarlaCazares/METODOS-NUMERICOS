#### Podemos decir:

\left(\begin{array}{ccc} l_{11}&0&0\\l_{21}&l_{22}&0\\l_{31}&l_{32}&l_{33} \end{array}\right)

#### Realizamos el producto:
$$\left(\begin{array}{ccc} 4&6&10\\6&25&19\\10&19&62 \end{array}\right)=\left(\begin{array}{ccc} l_{11}^{2}&l_{11}
l_{21}&l_{11}l_{31}\\l_{11}l_{21}&l_{21}^{2}+l_{22}^{2}&l_{21}l_{31}+l_{22}l_{32}\\l_{11}l_{31}&l_{31}l_{21}+l_{32}
l_{22}&l_{31}^{2}+l_{32}^{2}+l_{33}^{2} \end{array}\right)$$

#### Igualamos cada elemento a elemento

#### Fila $\Rightarrow 1:$
$$\begin{align}l_{11}^{2}=4 && \Rightarrow\ &&
l_{11}=2\end{align}$$

$$\begin{align}l_{11}l_{21}=6 && \Rightarrow\ && l_{21}=3\end{align}$$

$$\begin{align}l_{11}l_{31}=10 && \Rightarrow\ && l_{31}=5\end{align}$$

#### Fila $\Rightarrow 2:$
$$\begin{align}l_{21}^{2}l_{22}^{2}=25 && \Rightarrow \ && l_{22}=4\end{align}$$

$$\begin{align}l_{21}l_{31}+l_{22}l_{32}=19 && \Rightarrow\ && l_{32}=1\end{align}$$

#### Fila $\Rightarrow 3:$
$$\begin{align}l_{31}^{2}+l_{32}^{2}+l_{33}^{2}=62 && \Rightarrow\ && l_{33}=6\end{align}$$

#### La matriz $L$ es:
$$L=\left(\begin{array}{ccc} 2&0&0\\3&4&0\\5&1&6 \end{array}\right)$$

#### La descomposicin de cholesky es:
$$\left(\begin{array}{ccc} 4&6&10\\6&25&19\\10&19&62 \end{array}\right)=
\left(\begin{array}{ccc} 2&0&0\\3&4&0\\5&1&6 \end{array}\right)
\left(\begin{array}{ccc} 2&3&5\\0&4&1\\0&0&6 \end{array}\right)$$

#### Cdigo

```python
import numpy as np
def cholesky(A):
    A=np.array(A,float)
    L=np.zeros_like(A)
    n,_=np.shape(A)
    for j in range (n):
        for i in range(j,n):
            if i==j:
                L[i,j]=np.sqrt(A[i,j]-np.sum(L[i,:j]**2))
            else:
                L[i,j]=(A[i,j]-np.sum(L[i,:j]*L[j,:j]))/L[j,j]
    return L
A=[[4,6,10],[6,25,19],[10,19,62]]
l=cholesky(A)
print("La matriz L es:")
print(l)

```

### 4. Programe un algortmo para encontrar la norma de Frobenius para una matrz cuadrada de cualquier dimensin.

#### Dada una matrz $A \in \mathcal{M}_{m \times n}(\mathbb{C})$, su norma de Frobenius (llamada tambin la norma de Hilbert-Schmidt) se define como:
$$
\|A\|_{F}:=\left(\operatorname{tr}\left(A^{*} A\right)\right)^{1 / 2}
$$
#### O, de manera equivalente, como:

$$
\|A\|_{F}=\left(\sum_{j=1}^{m} \sum_{k=1}^{n}\left|A_{j, k}\right|^{2}\right)^{1 / 2}
$$

#### Vamos a desarrollar la siguiente matrz 4x4

$$
A=\left(\begin{array}{cccc}
5 & 2 & 2 & 0 \\
2 & 6 & -1 & -4 \\
-2 & 2 & 4 & -1 \\
1 & -1 & -5 & 2
\end{array}\right)
$$

#### Desarrollo

$\|A\|_{F}=\sqrt{(5)^{2}+(2)^{2}+(-2)^{2}+(1)^{2}+(2)^{2}+(6)^{2}+(2)^{2}+(-1)^{2}+(2)^{2}+(-1)^{2}+(4)^{2}+(-5)^{2}+(0)^{2}+(-4)^{2}+(-1)^{2}+(2)^{2}}$

$\|A\|_{F}=\sqrt{(25)+(4)+(4)+(1)+(4)+(36)+(4)+(1)+(4)+(1)+(16)+(25)+(0)+(16)+(1)+(4)}$

$\|A\|_{F}=\sqrt{146}$

$\|A\|_{F}= 12,0830$

### Cdigo

```python
import numpy as np
# MATRIZ A REALIZAR (A):
from math import sqrt
A=np.array([[5,2,-2,1],[2,6,2,-1],[2,-1,4,-5],[0,-4,-1,2]])
m=A.shape[0]
n=A.shape[1]
sum_raiz=0
for i in range (m):
    for j in range (n):
        sum_raiz+=pow(A[i][j],2)
        Res=sqrt(sum_raiz)
print("El resultado obtenido es:")
print(Res)        



```
